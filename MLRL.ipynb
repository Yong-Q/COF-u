```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 从csv文件中读取数据
df1 = pd.read_csv('./csv/RandomForestRegressor()_train.csv')
df1['Actual'] = df1['Actual'].astype(float)
df1['Prediction'] = df1['Prediction'].astype(float)
train_y = np.array(df1['Actual']).astype(float)
predictions = np.array(df1['Prediction']).astype(float)
train_y = df1['Actual']
predictions = df1['Prediction']

# 创建图形和坐标轴
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()
ax.fill_between([-4, 4], [-4.5, 3.5], [-3.5, 4.5], color='#ffdee4')

# 设置字体和标签
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 11
plt.ylabel('ML predicted $\mathregular{\Delta G_H}$ (eV)', fontsize=15)
plt.xlabel('DFT calculated $\mathregular{\Delta G_H}$ (eV)', fontsize=15)

# 绘制散点图
ax.scatter(train_y, predictions, color='#525b88', edgecolor='#dee2fc', s=40, linewidth='0.5')

# 添加文本标签
plt.text(1, -3, str('$\mathregular{MAE:}$ ') + str(0.12) + str('$\mathregular{\pm}$') + str(0.12), horizontalalignment='left', size='medium', color='black')
plt.text(1, -3.5, str('$\mathregular{R^2:}$ ') + str(round(0.12, 3)) + str('$\mathregular{\pm}$') + str(abs(round(0.12, 3))), horizontalalignment='left', size='medium', color='black')

# 设置坐标轴范围
lims = [-4, 4]
plt.xlim(lims)
plt.ylim(lims)

# 绘制对角线
plt.plot(lims, lims, '--', color='black')

# 保存图像
name = 'model.png'
#plt.savefig(name, dpi=600.0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 从csv文件中读取数据
df1 = pd.read_csv('./csv/RandomForestRegressor()_train.csv')
df1['Actual'] = df1['Actual'].astype(float)
df1['Prediction'] = df1['Prediction'].astype(float)
train_y = np.array(df1['Actual']).astype(np.float32)
predictions = np.array(df1['Prediction']).astype(np.float32)

train_y = df1['Actual']
predictions = df1['Prediction']

# 创建图形和坐标轴
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()


# 计算填充区域的边界坐标

ax.fill_between([0, 20.5], [-0.5, 19.5], [0.5, 20.5], color='#ffdee4')

# 设置字体和标签
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 11
plt.ylabel('ML predicted $\mathregular{\Delta G_H}$ (eV)', fontsize=15)
plt.xlabel('DFT calculated $\mathregular{\Delta G_H}$ (eV)', fontsize=15)
ax.scatter(train_y, predictions, color='#525b88', edgecolor='#dee2fc', s=40)
```

```python
import matplotlib.patches as mpatches
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df1 = pd.read_csv('./csv/RandomForestRegressor()_.csv')
df1['Actual'] = df1['Actual'].astype(float)
df1['Prediction'] = df1['Prediction'].astype(float)
test_y = np.array(df1['Actual']).astype(np.float32)
test_y_pred = np.array(df1['Prediction']).astype(np.float32)
fig = plt.figure( figsize = (7,5))
ax = fig.add_subplot()
lims = [0, 8]
plt.xlim(lims)
plt.ylim(lims)
ax.fill_between([0, 8], [-0.5, 7.5], [0.5, 8.5], color='#ffdee4')
#ax.fill_between([-4,4], [-4.5,3.5], [-3.5,4.5], color = '#ffdee4')
#ax.fill_between([-4,4], [-4.5,3.5], [-3.5,4.5], color = '#ffdee4')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 11.5
plt.ylabel('ML predicted (Score)', fontsize=19)
plt.xlabel('CDFT calculated (Score)', fontsize=19)
ax.scatter(train_y, predictions, color='#00bfa5', edgecolor='#dee2fc', s=40)
ax.scatter(test_y, test_y_pred, color = '#525b88', edgecolor='#dee2fc', s=40 )
maertrain=0.045
maertest=0.089
r2train=0.757
r2test=0.806
model="RFxx"
plt.text(5.5, 1.0, str('$\mathregular{MAE:}$ ') + str(abs(round(maertrain,3))) + '/' + str(abs(round(maertest,3))), horizontalalignment='left', size='medium', color='black')
plt.text(5.7, 0.5, str('$\mathregular{R^2:}$ ') + str(round(r2train,3)) + '/' + str(round(r2test,3)), horizontalalignment='left', size='medium', color='black')
name = str(model) + '.png'
#lims = [-4,4]

plt.plot(lims, lims, '--', color = 'black')
train_patch = mpatches.Patch(color='#00bfa5', label='Train')
test_patch = mpatches.Patch(color='#525b88', label='Test')
plt.legend(handles=[train_patch, test_patch])
plt.savefig('RF train-test 7*4.5.png', dpi=600, transparent=True)
```

```python
import matplotlib.patches as mpatches
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df1 = pd.read_csv('./csv/GradientBoostingRegressor()_.csv')
df1['Actual'] = df1['Actual'].astype(float)
df1['Prediction'] = df1['Prediction'].astype(float)
test_y = np.array(df1['Actual']).astype(np.float32)
test_y_pred = np.array(df1['Prediction']).astype(np.float32)
fig = plt.figure( figsize = (7,5))
ax = fig.add_subplot()
lims = [0, 8]
plt.xlim(lims)
plt.ylim(lims)
ax.fill_between([0, 8], [-0.5, 7.5], [0.5, 8.5], color='#ffdee4')
#ax.fill_between([-4,4], [-4.5,3.5], [-3.5,4.5], color = '#ffdee4')
#ax.fill_between([-4,4], [-4.5,3.5], [-3.5,4.5], color = '#ffdee4')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 11.5
plt.ylabel('ML predicted (Score)', fontsize=19)
plt.xlabel('CDFT calculated (Score)', fontsize=19)
ax.scatter(train_y, predictions, color='#00bfa5', edgecolor='#dee2fc', s=40)
ax.scatter(test_y, test_y_pred, color = '#525b88', edgecolor='#dee2fc', s=40 )
maertrain=0.042
maertest=0.071
r2train=0.757
r2test=0.741
model="RFxx"
plt.text(5.5, 1.0, str('$\mathregular{MAE:}$ ') + str(abs(round(maertrain,3))) + '/' + str(abs(round(maertest,3))), horizontalalignment='left', size='medium', color='black')
plt.text(5.7, 0.5, str('$\mathregular{R^2:}$ ') + str(round(r2train,3)) + '/' + str(round(r2test,3)), horizontalalignment='left', size='medium', color='black')
name = str(model) + '.png'
#lims = [-4,4]

plt.plot(lims, lims, '--', color = 'black')
train_patch = mpatches.Patch(color='#00bfa5', label='Train')
test_patch = mpatches.Patch(color='#525b88', label='Test')
plt.legend(handles=[train_patch, test_patch])
plt.savefig('GBR train-test 7*4.5.png', dpi=600, transparent=True)
```
